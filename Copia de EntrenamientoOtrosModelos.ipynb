```
def modeloVgg16():
  vgg = applications.vgg16.VGG16()
  cnn = Sequential()
  for capa in vgg.layers:
    cnn.add(capa)
  cnn.layers.pop()
  for layer in cnn.layers:
    layer.trainable=False
  cnn.add(Dense(6,activation='softmax'))
  return cnn
```

```
nuevoModelo = modeloVgg16()
nuevoModelo.summary() 
```
